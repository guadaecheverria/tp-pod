# Cancelaciones y Delays de vuelos del 2015

Creese una cuenta en Kaggle e importe los archivos del dataset del siguiente link: https://www.kaggle.com/datasets/usdot/flight-delays. Cree los dataframes `airlines`, `airports`, y `flights` apartir de esos archivos.

```python
airlines = pd.read_csv('airlines.csv')
airports = pd.read_csv('airports.csv')
flights = pd.read_csv('flights.csv')

print("Airlines DataFrame:",airlines.head())
print("Airports DataFrame:",airports.head())
print("Flights DataFrame:",flights.head())
```

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
# join entre flights y airlines usando 'AIRLINE'
flights_airlines = pd.merge(flights, airlines, how='left', left_on='AIRLINE', right_on='AIRLINE')

# join entre flights_airlines y airports para el aeropuerto de origen (Origin)
flights_airports_origin = pd.merge(flights_airlines, airports, how='left', left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', suffixes=('_origin', '_airport'))

# join entre flights_airports_origin y airports para el aeropuerto de destino (Destination)
flights_airports = pd.merge(flights_airports_origin, airports, how='left', left_on='DESTINATION_AIRPORT', right_on='IATA_CODE', suffixes=('_origin', '_destination'))

print(flights_airports.head())


```

¿Cuantos vuelos fueron al aeropuerto JFK?

```python
flights_to_jfk = flights_airports[flights_airports['DESTINATION_AIRPORT'] == 'JFK']
num_flights_to_jfk = flights_to_jfk.shape[0]

print(f"Cantidad de vuelos al aeropuerto JFK: {num_flights_to_jfk}")
```

¿Cuantos vuelos hizo la aerolinea AA?

```python
flights_aa = flights_airports[flights_airports['AIRLINE'] == 'AA']
num_flights_aa = flights_aa.shape[0]

print(f"Cantidad de vuelos realizados por la aerolínea AA: {num_flights_aa}")
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
flights_with_delays = flights_airports[flights_airports['ARRIVAL_DELAY'] > 0]

delay_counts = flights_with_delays.groupby('AIRLINE').size()

top_5_airlines_with_least_delays = delay_counts.nsmallest(5)

print("Las 5 aerolíneas con menos vuelos con atrasos son:")
print(top_5_airlines_with_least_delays)
```

¿Que aerolineas (las primeras 5) tuvo la mayor cantidad de vuelos con atrasos? Imprimirlas.

```python
flights_with_delays = flights_airports[flights_airports['ARRIVAL_DELAY'] > 0]

delay_counts = flights_with_delays.groupby('AIRLINE').size()

top_5_airlines_with_most_delays = delay_counts.nlargest(5)

print("Las 5 aerolíneas con más vuelos con atrasos son:")
print(top_5_airlines_with_most_delays)
```

Compruebe si hay columnas con celdas vacias.

```python
flights_airlines = pd.merge(flights, airlines, how='left', left_on='AIRLINE', right_on='IATA_CODE')

flights_airlines_airports = pd.merge(flights_airlines, airports, how='left', left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', suffixes=('_origin', '_destination'))

missing_values = flights_airlines_airports.isnull().sum()
print("Celdas vacías por columna:")
print(missing_values)

missing_columns = missing_values[missing_values > 0]
print("\nColumnas con celdas vacías:")
print(missing_columns)
```

Haga una imputación de datos COMPLETA del dataframe. Pueden escojer cualquier estrategia y no necesariamente todas las columnas deben seguir la misma estrategia.

```python
print(flights_airlines_airports.dtypes)

numeric_columns = flights_airlines_airports.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_columns:
    flights_airlines_airports[col].fillna(flights_airlines_airports[col].mean(), inplace=True)

categorical_columns = flights_airlines_airports.select_dtypes(include=['object']).columns
for col in categorical_columns:
    flights_airlines_airports[col].fillna(flights_airlines_airports[col].mode()[0], inplace=True)

missing_values_after_imputation = flights_airlines_airports.isnull().sum()
print("\nCeldas vacías después de la imputación:")
print(missing_values_after_imputation)
```
