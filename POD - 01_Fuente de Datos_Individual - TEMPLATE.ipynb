¿Quien es el autor y titulo del libro? Responderlo usando las herramientas utilizadas en la celda anterior (Tip: abrir el archivo de texto y ver como son las primeras lineas).

```python
with open('pg2000.txt', 'r', encoding='utf-8') as archivo:
    lineas = archivo.readlines()

titulo = None
autor = None
for linea in lineas:
    if linea.startswith("Title:"):
        titulo = linea.replace("Title:", "").strip()
    elif linea.startswith("Author:"):
        autor = linea.replace("Author:", "").strip()
    if titulo and autor:
        break

print("Título del libro:", titulo)
print("Autor del libro:", autor)
```

## Consignas

¿Como luce la base de datos? Imprimirla.


```python
conn = sqlite3.connect('db.db')
cur = conn.cursor()

cur.execute("CREATE TABLE IF NOT EXISTS t (GameNumber TEXT, GameLength TEXT);")
with open('snakes_count_10000.csv', 'r') as csv_file:
    lines = csv_file.readlines()[1:]
    lines = [l.split(',') for l in lines]
    to_db = [(l[0].strip(), l[1].strip()) for l in lines]


cur.executemany("INSERT INTO t (GameNumber, GameLength) VALUES (?, ?);", to_db)
conn.commit()

cur.execute("SELECT * FROM t LIMIT 10;")
rows = cur.fetchall()

print("Primeros 10 registros de la base de datos:")
for row in rows:
    print(row)



```

¿Cuantas filas hay en la base de datos?

```python
cur.execute("SELECT COUNT(*) FROM t;")
total_filas = cur.fetchone()[0]

print(f"El número total de filas en la base de datos es: {total_filas}")

```

¿Cual es el máximo `GameLength` y en qué `GameNumber` se realizó?

```python
cur.execute("SELECT GameNumber, GameLength FROM t ORDER BY GameLength DESC LIMIT 1;")
resultado = cur.fetchone()

print(f"El máximo GameLength es: {resultado[1]}, y ocurrió en el GameNumber: {resultado[0]}")


```

Hacer una request GET e imprimir uno de los campos de la respuesta:

```python
import requests

url = 'https://catfact.ninja/facts'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(data['data'][0]['fact'])
else:
    print(f"Error: {response.status_code}")


```
